# EDA (Exploratory Data Analysis)
EDA is a method to analyze and summarize the dataset. In python we can use Pandas library to perform EDA on the dataset.

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset 
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
# now we've to summarize the dataset also called summary statistics
df.describe()
# in this we'll get the summary of numeric values
```

```python
# finding null values 
df.isnull()
```

```python
df.isnull().sum()
# the way in which we deal with null values is called data wrangling 
# in EDA only we explore these values
```

```python
# now we're checking it's percentage 
df.isnull().mean()*100
```

# unique value of any variable is more than 10 there are 90% cances that it would be a numeric.

```python
# finding unique values 
df.nunique()
```

```python
df.columns
# here are the columns names that we see the names of columns 
```

```python
# visualize the nan valuesin heatmap
sns.heatmap(df.isnull())
```

```python
# above is'nt clear so we've to use this formula 
sns.heatmap(df.isnull(), yticklabels=False,cbar=False, cmap='viridis')
```

```python
# now we've to remove deck 
df.drop(columns='deck', inplace=True)
```

```python
df.columns
```

# in countplot we can't give values for y -axis. It will count it'self

```python
# visualize the data
sns.countplot(data=df,x='survived')
```

```python
# barplot 
sns.barplot(data=df,x='survived',y='age',hue='who')
```

```python
sns.countplot(data=df,x='pclass',palette="viridis")
# on barplot if we're cheking count we've to give only categorical variabes on one side 
# on x-axis give categorical variable and on y-axis give numerical variables
```

```python
# now we're checking the data distribution 
sns.histplot(data=df,x="age", bins=15)
# this hist plot describe tat how many age of people are there in ship 
```

```python
# we can also see data distribution by this way '
df['age'].hist(bins=15, alpha=0.7, color="green")
# hist plot always made for a numeric values 
```

# Can we make a boxplot for two numeric variables ?

```python
#  now we're making a boxplot 
sns.boxplot(data=df,y="age",x="survived",hue="who",palette="viridis")
# while makig a box plot we have to take numeric value on one side and take categorical value on other axis 
# in men there is the outlier that we'll remove in the data wrangling 
```

```python
# if we've to make graph for two numeric values we'll make scater values 
sns.scatterplot(x="age",y="fare",data=df, hue="pclass")
plt.ylim(0,300)
# see this data is'nt normalize and this we'll do in the datawrangling part
```

```python
# now we're going to make pair plot 
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="survived")
```

```python
# stacked plot 
df.groupby(['pclass','who'])['fare'].mean().unstack().plot(kind='bar',stacked=True)
```

```python
# swarmplot
sns.swarmplot(data=df,x='age',y='fare')
# therr is no much difference between scatter and swarm plot 
# if we draw swarmplot on bar plot it began to show something 
```

```python
sns.swarmplot(data=df,x='survived',y='fare',hue='who')
plt.ylim(0,300)
```

```python
sns.pointplot(x='pclass',y='fare',hue='embarked',data=df)
```

```python
# relationship of two numeric variables
sns.lmplot(data=df,y='fare',x='age')
plt.ylim(0,300)
```

```python
sns.lmplot(data=df[df['fare'] < 300], y='fare', x='age')
# in this we use condition that shw less than 300 fare show plot 
```

```python
# how to adda column and make a new age group 
# in previous we've made 
sns.histplot(data=df,x='age')
```

```python
# grouping 
df.columns
```

```python
df['age_group'] = pd.cut(df['age'],[0,10,18,30,50,80],labels=['kids', 'children','youth','middle aged','old'])
```

```python
df.head()
```

```python
df.columns
```

```python
df.groupby("age_group")['survived'].mean().plot(kind='bar')
```

```python
# jaisa hm chahty hai wiasa bhi pair plot bna skty hai by using vars= command 
 
sns.pairplot(df,hue='survived',vars=['age','fare','pclass'])
```

```python
# assignment add a family size column 
df['family_size'] = pd.cut(df['fare'],[0,100,200,300,400,500,600],labels=['lower','lower middle','upper','upper middle','elite','standard'])
```
