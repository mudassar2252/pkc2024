# 16- Spliting  a string into multiple columns 
this is used when we have to give  one or more than one ways of dataframe  

```python
# exammple 
# import libraries

import pandas as pd

# create a dataframe

df = pd.DataFrame({'name': ['John smith', 'Omkar Pittala', 'Sahn Mukh'],
                   'location': ['New York, USA', 'Los Angeles, USA', 'Chicago, USA']})

df
```

```python
df.name.str.split(' ')
```

```python
df.location.str.split(' ')
```

```python
# if we put expand=true 
df.name.str.split(' ',expand=True)
```

```python
df.location.str.split(' ',expand=True)
```

```python
# adding the new column in column 
df[['first_name', 'last_name']] = df.name.str.split(' ',expand=True)
df
```

```python
df[['city_name','state_of']] = df.location.str.split(', ',expand=True)
df
```

```python
# now we have to refine the data manipulation
df = df[['first_name', 'last_name' , 'city_name', 'state_of']] 
df

```

# 23- Reshaping a dataframe 

```python
# crrat a simple dataframe 

df = pd.DataFrame({'A': ['foo', 'bar', 'foo', 'bar',
                         'foo', 'bar', 'foo', 'foo'],
                   'B': ['one', 'one', 'two', 'three',
                         'two', 'two', 'one', 'three'],
                   'C': np.random.randn(8),
                   'D': np.random.randn(8)})
df.head()
```

```python
# now i want to reshape thisdataframe use mmelt formula
# this is 
#   A  B  C  D  E

df.melt(id_vars=['A', 'B'], value_vars=['C', 'D'])



```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame
data = {'Product': ['A', 'B', 'C', 'D'], 'Sales': [500, 300, 450, 600]}
df = pd.DataFrame(data)

# Create a bar plot
sns.barplot(x='Product', y='Sales', data=df, palette='Blues')

# Add titles and labels
plt.title("Product Sales")
plt.xlabel("Products")
plt.ylabel("Sales (in $)")

# Show the plot
plt.show()
```
