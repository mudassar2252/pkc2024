# K-Means Clustering

```python
# ML mei clustering ka ye mtlb hota hai kah agr  aik data mei three colors ky points bn rhy hai to jb hm isko m sy pass krty hai to wo three alg alg apny group bna lety hai 
# is mei kafi theoratical aur mathematical equations use hoti hai lakin hm yhan yehi deikhen gy kah hm isko kaisy X aur y ko define krty hai 
# is mei label nhi hoty mtlb is mei hmny sirf X pr kam krna hai 
# last lecture mei hmny X aur y mei define kia tha lakin is mei sirf X ko krna hai 
# this is called unsupervised machine learning  
```

```python
# import libraries and dataset 
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
df = sns.load_dataset("titanic")
df.head()
```

```python
# ab hm model ko ly kr aty hai 
from sklearn.cluster import KMeans
# sath mei hmstandard scaller ko ly kr aen gy ta kah data normalize ho sky 
# data ki main sense bhi normalization ky bad hi bnti hai 
from sklearn.preprocessing import StandardScaler 
# aur idhr hm X ko ly lety hai 
# importing tha data in features 
X = df[['age','fare']]
X.age.fillna(value=X['age'].mean(),inplace=True)

# scaled data 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# kmeans model ko create kr ty hai
model = KMeans(n_clusters=5)
model.fit(X_scaled)
# now we've to assign the cluster  label to dataset 
df['cluster'] = model.labels_
```

```python
# ab hum iska scatter plot bna kr deikhty hai 
plt.scatter(df['age'],df['fare'])
plt.show()
```

```python
plt.scatter(df['age'],df['fare'],c=df['cluster'],cmap='rainbow')
# hm centeroid bhi bna skty hai isky 
```
