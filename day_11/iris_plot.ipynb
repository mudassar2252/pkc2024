```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
iris.columns
```

```python
iris.info()
```

```python
sns.barplot(data=iris , x="petal_width", y="petal_length", hue="species",palette="ocean")
```

```python
sns.barplot(data=iris , x="species", y="sepal_length",palette="viridis")
plt.title("Species")

plt.show()
```

```python
iris.hist()
```

```python
sns.pairplot(data=iris)
```

```python
sns.barplot(data=iris, x="petal_length" , y="petal_width",hue="species", palette="viridis")
plt.title("Species distribution")
plt.xlabel("petal length")
plt.ylabel("petal width")

plt.show()
```

```python
sns.scatterplot(data=iris, x="petal_length" , y="petal_width",hue="species", palette="viridis")
plt.title("Species distribution")
plt.xlabel("petal length")
plt.ylabel("petal width")

plt.show()
```

```python
sns.boxenplot(data=iris, x="petal_length" , y="petal_width",hue="species", palette="viridis")
plt.title("Species distribution")
plt.xlabel("petal length")
plt.ylabel("petal width")

plt.show()
```

```python
sns.lineplot(data=iris, x="petal_length" , y="petal_width",hue="species", palette="viridis")
plt.title("Species distribution")
plt.xlabel("petal length")
plt.ylabel("petal width")

plt.show()
```

```python
sns.countplot(data=iris, x="species" ,  palette="husl")
plt.title("Species distribution")

plt.show() 
```

```python
sns.stripplot(data=iris, x="sepal_width" , y="sepal_length",hue="species", palette="husl")
plt.title("Species distribution")
plt.xlabel("sepal width")
plt.ylabel("sepal length")

plt.show()
```

```python
sns.boxplot(data=iris, x="sepal_width" , y="sepal_length",hue="species", palette="husl")
plt.title("Species distribution")
plt.xlabel("sepal length")
plt.ylabel("sepal width")

plt.show()
```

```python
sns.scatterplot(data=iris, x="petal_length", y="petal_width", hue="species", palette="husl")
plt.title("Petal Length vs Petal Width")
plt.show()
```

```python
sns.violinplot(data=iris, x="species", y="sepal_length", palette="husl")
plt.title("Sepal Length")
plt.show()
```
